```python
from os import walk
f = []
for (dirpath, dirnames, filenames) in walk(mypath):
    for file in filenames:
        tempPath = dirpath + '\\' + file
        
        f.append(tempPath)
    
```

```python
images = []
label = []
for file in f:
    label.append(file.split('_')[0][-1])
    images.append(cv2.imread(file, cv2.IMREAD_GRAYSCALE))
```

```python
#Data augmentation:
from keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
datagen = ImageDataGenerator(shear_range=0.1, zoom_range=0.1,width_shift_range=0.1,height_shift_range=0.1, fill_mode='constant', cval=0) 
xx=0
for file in f:
    
    fileformat = file.split('\\')
    img = load_img(file,color_mode="grayscale")  
    x = img_to_array(img) 
    
    #Reshape the input image 
    x = x.reshape((1, ) + x.shape)  
    i = 0

    # generate 5 new augmented images 
    for batch in datagen.flow(x, batch_size = 1, 
                      save_to_dir ='temp3'+ '\\' + fileformat[-2]+'\\', #+ fileformat[-1][:-4],
                     save_format ='jpg'):
        i += 1
        if i > 10: 
            break
```
